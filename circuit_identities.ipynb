```python
import numpy as np
from qiskit import(
  QuantumCircuit,
  execute,
  Aer)
from math import pi

from qiskit.visualization import plot_bloch_multivector, plot_histogram
from qiskit_textbook.tools import array_to_latex


simulator = Aer.get_backend('qasm_simulator')
statevector_backend = Aer.get_backend('statevector_simulator')
unitary_backend = Aer.get_backend('unitary_simulator')

qc = QuantumCircuit(2)
qc.h(1)
qc.cx(0, 1)
qc.h(1)

unitary = execute(qc, unitary_backend).result().get_unitary()
array_to_latex(unitary, pretext="\\text{Circuit = }\n")
# print(unitary)
```

# Controlled-H (CH)
Applies H only if control qubit is set to |1>

```python

qc = QuantumCircuit(2)

qc.x(0)

qc.ry(pi/4, 1)
qc.cx(0, 1)
qc.ry(-pi/4, 1)

state = execute(qc, statevector_backend).result().get_statevector()
plot_bloch_multivector(state)
```

```python
# swap the states of |+> and |->

qc = QuantumCircuit(2)

# qc.ry(0, 0)
# qc.ry(pi, 1)

qc.cz(0, 1)
qc.cz(1, 0)
# qc.cz(0, 1)
# qc.cy(0, 1)

unitary = execute(qc, unitary_backend).result().get_unitary()

state = execute(qc, statevector_backend).result().get_statevector()
array_to_latex(unitary, pretext="\\text{Circuit = }\n")
plot_bloch_multivector(state)
```

# Controlled Rotations


```python
c = 0 # control
t = 1 # target

qc = QuantumCircuit(2)

theta = pi # theta can be anything (pi chosen arbitrarily)
qc.x(c)
qc.ry(theta/2, t)
qc.cx(c,t)
qc.ry(-theta/2,t)
qc.cx(c,t)

# qc.draw()
state = execute(qc, statevector_backend).result().get_statevector()
plot_bloch_multivector(state)
```

# Arbitrary Rotations from H and T


```python
qc = QuantumCircuit(1)
qc.h(0)
qc.t(0)
qc.h(0)
qc.t(0)

state = execute(qc, statevector_backend).result().get_statevector()
plot_bloch_multivector(state)
```
