```python
import numpy as np
from qiskit import(
  QuantumCircuit,
  execute,
  Aer)
from math import pi

from qiskit.visualization import plot_bloch_multivector, plot_histogram

# Use Aer's qasm_simulator
simulator = Aer.get_backend('qasm_simulator')
statevector_backend = Aer.get_backend('statevector_simulator')

qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.cu1(pi/4, 0, 1)
display(qc.draw())


final_state = execute(qc, statevector_backend).result().get_statevector()
plot_bloch_multivector(final_state)
```

```python
qc = QuantumCircuit(1)

# qc.h(0)

# u1 = rotatie in jurul la Z
# u2 = face o rotatie de pi /2 in jurul la X dupa care in jurul la Z 
# u3 = rotatie in jurl la X la Z Y 

qc.u2(0, pi/2, 0) # rotatie in jurul la Z

display(qc.draw())

final_state = execute(qc, statevector_backend).result().get_statevector()
plot_bloch_multivector(final_state)
```
