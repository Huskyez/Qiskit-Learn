# Quantum Teleportation


```python
import numpy as np
from math import sqrt
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, execute, Aer, IBMQ
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.extensions import Initialize
from qiskit_textbook.tools import random_state, array_to_latex


unitary_backend = Aer.get_backend("unitary_simulator")
statevector_backend = Aer.get_backend("statevector_simulator")


def init_state(qc, qubit, state=None):
    """
    Initializes the qubit in qc to some state, if the state is not specified it will be random
    Returns the inverse of the state
    """
    if state == None:
        state = random_state(1)
    init_gate = Initialize(state)
    init_gate.label = "|q>"
    qc.append(init_gate, [qubit])
    return init_gate.gates_to_uncompute()

def create_bell_pair(qc, a, b):
    """Creates a bell pair in qc using qubits a & b"""
    qc.h(a) # Put qubit a into state |+>
    qc.cx(a,b) # CNOT with a as control and b as target

def alice_gates(qc, psi, a):
    qc.cx(psi, a)
    qc.h(psi)
    
def measure_and_send(qc, a, b):
    """Measures qubits a & b and 'sends' the results to Bob"""
    qc.barrier()
    qc.measure(a,0)
    qc.measure(b,1)
    
def bob_gates(qc, a, b, c):
    qc.cz(a, c)
    qc.cx(b, c)
    
    
 # Protocol uses 3 qubits and 2 classical bits in 2 different registers
qr = QuantumRegister(3, name="q")
result = ClassicalRegister(1, name="r")
teleportation_circuit = QuantumCircuit(qr, result)

## STEP 0 (initialize Alice's qubit)
inverse = init_state(teleportation_circuit, 0, [sqrt(3)/2, 1/2])
teleportation_circuit.barrier()

## STEP 1
create_bell_pair(teleportation_circuit, 1, 2)

## STEP 2
teleportation_circuit.barrier() # Use barrier to separate steps
alice_gates(teleportation_circuit, 0, 1)

## STEP 3
# measure_and_send(teleportation_circuit, 0, 1)

## STEP 4
# teleportation_circuit.barrier() # Use barrier to separate steps
bob_gates(teleportation_circuit, 0, 1, 2)

display(teleportation_circuit.draw())

# unitary = execute(teleportation_circuit, unitary_backend).result().get_unitary()
# array_to_latex(unitary)
```

### Represent Bob's state at the end of the circuit

```python
state = execute(teleportation_circuit, statevector_backend).result().get_statevector()
plot_bloch_multivector(state)
```

### Verify Bob's state


```python
## STEP 5 (verify that Bob's qubit is in state |q>)
teleportation_circuit.barrier()
teleportation_circuit.append(inverse, [2])
teleportation_circuit.measure(2, 0)

display(teleportation_circuit.draw())
```

### Test the circuit on the simulator

```python
backend = Aer.get_backend("qasm_simulator")
counts = execute(teleportation_circuit, backend, shots=1024).result().get_counts()
plot_histogram(counts)
```

We got 0, 100% of the time, which is what we expected.
